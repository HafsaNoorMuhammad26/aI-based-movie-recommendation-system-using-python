```python
#dataset
#create in tuples and in each tuple:(Movie_name, Genre, Rating, Min_age)
#age group=0 means for all ages, age group=13 means teen agers and for 18 adults only

movies = [
    ("Inception", "Sci-Fi", 8.8, 13),
    ("Intersteller", "Sci-Fi", 8.6, 13),
    ("The Martian", "Sci-Fi", 8.0, 13),
    ("Avengers: Endgame", "Action", 8.4, 16),
    ("John Wick", "Action", 7.4, 18),
    ("Spider-Man: No Way Home", "Action", 8.2, 10),
    ("The Lion King", "Animation", 8.5, 0),
    ("The Story", "Animation", 7.5, 0),
    ("Frozen", "Animation", 7.5, 0),
    ("The Dark Knight", "Action", 9.0, 13),
    ("Arrival", "Sci-Fi", 7.9, 13),
    ("Kung Fu Panda", "Animation", 7.6, 0)
]
```

```python
valid_genres = ["Action", "Sci-Fi", "Animation"]

#print welcome message
print("=" * 50)
print("Welcome to the AI-Based Movie Recommendation System")
print("=" * 50)
print()

#system takes user name as input
name = input("Enter your name: ").strip()
while name == "" or name.isdigit():
  print("Error: Please enter a valid name. ")
  name = input("Enter your name: ").strip()

#system takes user age as input
age = input("Enter your age: ").strip()
while True:
  if not age.isdigit():
    print("Error: Please enter a valid age (it must be a number). ")
  elif int(age) <= 0:
    print("Error: Please enter a valid age (it must be greater than 0). ")
  elif int(age) > 120:
    print("Error: Please enter a realistic age (1 - 120). ")
  else:
    break
  age = input("Enter your age: ").strip()

age = int(age)

#system takes genre as input
print()
print("Available Genres: Action, Sci-Fi, Animation")
genre = input("Enter your preferred genre: ").strip().title()
while genre not in valid_genres:
  print("Error: Please enter a valid genre. ")
  genre = input("Enter your preferred genre: ").strip().title

#system takes user mood as input
print()
valid_moods = ["happy", "adventurous", "relaxed"]
mood = input("How are you feeling today? (happy / adventurous / relaxed): ").strip().lower()
while mood not in valid_moods:
  print("Error: Please enter a valid mood. ")
  mood = input("How are you feeling today? (happy / adventurous / relaxed): ").strip().lower()

#system recommends movies based on user input
recommended = []
blocked_count  = 0  # tracks how many movies were blocked due to age

for movie in movies:
  movie_name = movie[0]
  movie_genre = movie[1]
  movie_rating = movie[2]
  min_age = movie[3]

 #condition 1: Genre must match what user selected
  if movie_genre.lower() != genre.lower():
    continue

  #condition 2: Block age-restricted movies for underage users
  if age < min_age:
    blocked_count += 1
    continue

  #condition 3: Only recommend movies rated 7.5 and above (high-rated priority)
  if movie_rating < 7.5:
     continue

  recommended.append(movie)

for i in range(len(recommended)):
  for j in range(i+1, len(recommended)):
    if recommended[j][2] > recommended[i][2]:
      recommended[i], recommended[j] = recommended[j], recommended[i]

mood_note = ""
if mood == "adventurous":
  mode_note = "Great pick for an adventurous mood! "
elif mood == "happy":
  mood_note = "Great pick for a happy mood! "
elif mood == "relaxed":
  mood_note = "Great pick for a relaxed mood! "
else:
  mood_note = "Great pick, enjoy your movie night! "

print()
print("=" * 50)
print(f"Top Picks for {name} ({genre})")
print("=" * 50)

if len(recommended) == 0:
  print("Sorry, no movies found for your preferences. ")
  print("Try a different genre or check back later! ")

else:
  print(f" {mood_note}")
  print()
  count = 1
  for movie in recommended:
    print(f" {count}. {movie[0]} | Rating: {movie[2]} | Age: {movie[3]}")
    count += 1

print()
print("=" * 50)
print(f"  Total movies recommended: {len(recommended)}")
print("  Thanks for using the system,", name + "!")
print("=" * 50)
```
